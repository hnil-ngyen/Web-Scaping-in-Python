```python
import requests
from bs4 import BeautifulSoup

# URL của trang khi tìm những bài báo liên quan lĩnh vực tài chính ngân hàng. 
url = "https://cafef.vn/tai-chinh-ngan-hang.chn"

# Gửi yêu cầu GET đến trang web và lấy nội dung.
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# Tìm các phần tử chứa thông tin bài báo.
articles = soup.find_all("div", class_="box-category-item")

# Đường link gốc của trang web
base_url = "https://cafef.vn"

# Lặp qua các bài báo và lấy thông tin cần thiết.
for article in articles:
    title = article.find("h3").text.strip()
    time = article.find("span", class_="time").text.strip()
    sapo = article.find("p", class_="sapo").text.strip()
    link = base_url + article.find("a")["href"] 
    
    print("Tên bài báo:", title)
    print("Thời gian đăng tải:", time)
    print("Nội dung chính:", sapo)
    print("Đường link bài báo:", link)
    print("\n")
```
